# Linuex常用命令  
1. 目录处理命令 ls
    1. ls  显示当前目录内的所有文件及目录（不包括隐藏）。  
    2. ls -a 显示所有文件及目录（包含隐藏）。
    3. ls -l 显示当前目录下的所有文件及目录，并显示细节 (如权限， 文件索引数， 所有者，用户组，大小，修改时间，)。  
    4. ls -d 显示本目录的信息（不加参数是显示本目录下的文件）。  
    5. ls -h 人性化显示，比如文件大小用kb等等显示而不是字节。  
    6. ls -i 显示文件的id索引（文件系统中每个文件或者目录的唯一标识）。  
2. 目录操作  
    1. mkdir 新建空目录，如果需要创建多级目录需要使用参数 -p递归创建。  
    2. cd  切换目录，"."表示当前目录，“..”表示上一级目录。  
    3. pwd 显示当前目录的绝对路径。  
    4. rmdir 删除一个空目录。  
    5. cp  拷贝文件, 如果要保存文件的信息，使用-p参数；如果拷贝目录使用-r参数。  
    6. mv 剪切文件（跟复制一样，可以在操作的同时改名）。  
    7. rm 删除文件，-r表示删除一个目录， -f强制删除而不询问。  
3. 文件操作  
    1. touch 创建一个文件，不建议使用空格作为文件名。  
    2. cat  显示文件内容（如果文件内容较多会显示最后一屏）， -n会添加行号； tac 倒叙显示文件内容。  
    3. more 分屏显示文件内容， 空格翻页，回车换行， q退出。  
    4. less 分屏显示文件内容， pgup和上箭头可以向上翻页，/检索内容 可以进行检索关键字，检索时n键翻页检索内容， 其他操作同more。  
    5. head/tail -n  数字   显示文件的前几行，不加参数显示10行，同时tail可以动态显示文件内容，文件的修改操作都会显示出来。 
4. 文件链接  
    1. ln -s 源文件  目标文件  
    生成软链接文件， 软链接相当于window的快捷方式，所有用户均有所有的权限，同时文件类型显示为l（link），软链接文件很小。  
    2. ln 源文件 目标文件  
    类似cp -p操作，但是硬链接可以同步更新（无论修改哪个文件都会同步）。硬链接具有跟源文件一致的i节点号，但是软链接不同。硬链接可以在源文件删除的情况下继续访问，而软链接不行。硬链接不能跨分区，不能针对目录使用。  
5. 权限管理chmod  
    只有所有者和root账号可以修改文件/目录的权限。  
    1. 使用+-=和ugoa来修改权限  
    例如： chmod g+w 文件/目录路径，对group用户增加写权限。a是all的意思，=代表不管之前是如何权限，更改为=之后的权限。  
    2. 常用的数字权限设置方法  
    r:4 w:2 x:1 所以777代表所有用户的所有权限，例如： chmod 777 路径。  
    3. 文件/目录权限的含义  
        对文件没有写权限，但是对文件所在的目录有写权限，那么就可以删除文件。
    4. -R选项  
    当时用-R选项时，递归的修改目录及他所包含的内容的权限。 

 
|命令|文件|目录|  
|:-:|:-:|:-:|  
|r|可以查看文件的内容|可以查看目录的内容(ls)|  
|w|可以编辑文件|可以对目录内的东西进行创建和删除|  
|x|可以执行文件|可以进入目录(cd)|  

6.  umask默认权限设置  
    当新建一个文件或目录时，会为它增添默认的权限，这个权限可以通过umsk -S命令查看，不建议修改这个默认设置  
7. chown更改拥有者；chgrp更改用户组  
    当使用ls -l查看文件属性时，会有拥有者和用户组的信息，可通过命令进行修改（只有root用户有权限，并且用户或者用户组必须存在）  
8. find查找命令  
    find会遍历制定的分区，使用较多系统资源
        find 分区 -name 文件名  ： 根据文件名检索分区，注意严格区分大小写，并且严格匹配文件名  
        find 分区 -iname 文件名 ： 不区分大小写，且可以用？或\*来匹配一个字符或任意个字符  
        find 分区 -size 分区 文件大小： 根据大小检索文件， + - = 分别代表大于，小于和等于。注意这里大小是块数，每块=0.5K  
        find 分区 -type d/f/l  ： 检索特定类型的文件，d目录，f文件，l链接文件  
        find 分区 -amin/-cmin/-mmin 数值 : 根据时间来检索，amin文件access时间， cmin修改权限时间， mmin编辑时间  
        find 分区 -inum  数值   ： 根据文件索引号来检索，可以查看文件是否有硬链接  
        find 分区 -user/-group 用户/用户组 : 根据用户（组）来查找  
        -a/-o 选项连接符 a是and，o是or  
        -exec/-o 命令 {} \; 可以根据检索到的结果，执行命令。-ok是询问是否执行  
9. locate查找  
    在一个文件索引库中查找文件，可以用uypdatedb来更新，速度更快。但是会忽略一些文件，比如/tmp下的文件。可以用来查找系统文件等等  
    which查找命令路径（并且可以查看命令是否有别名 ） whereis（可以查看命令路径和帮助文档的路径）
    grep -v/-i 关键词 文件路径 ： 在文件内容中搜索， -v 忽略特定条件比如^#， -i不区分大小写
